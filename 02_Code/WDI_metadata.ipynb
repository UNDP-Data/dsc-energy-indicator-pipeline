Step1: Select the sheet we want

```python
import pandas as pd
# CHANGE FILE PATH TO LOCATION OF YOUR WDIEXCEL.xlsx
file_path = '/Users/17mac/Desktop/WDIEXCEL.xlsx'

# Load the sheet named 'Data' and 'Series'
df_data = pd.read_excel(file_path, sheet_name='Data')
df_series = pd.read_excel(file_path, sheet_name='Series')
```

Step2: Extract metadata

```python
import re

series_dict = df_series.set_index('Series Code')['Topic'].to_dict()
series_dict_name = df_data.set_index('Indicator Code')['Indicator Name'].to_dict()
series_dict_definition = df_series.set_index('Series Code')['Long definition'].to_dict()
data_dict_source = df_series.set_index('Series Code')['Source'].to_dict()


# Indicator Code (sheet: Data)
data = df_data[['Indicator Code']]


# Category (sheet: Series)
def extract_category(indicator_code):
    topic = series_dict.get(indicator_code, '')
    return topic.split(':')[0] if ':' in topic else topic


# Indicator (sheet: Data)
def extract_indicator(indicator_code):
    name = series_dict_name.get(indicator_code, '')
    return name.split('(')[0].strip() if '(' in name else name.strip()


# Dimension (sheet: Series)
def extract_dimension(indicator_code):
    topic = series_dict.get(indicator_code, '')
    return topic.split(':')[1].strip() if ':' in topic else ''


# Units (sheet: Data)
def extract_units(indicator_code):
    name = series_dict_name.get(indicator_code, '')
    start = name.find('(')
    end = name.find(')')
    return name[start+1:end].strip() if start != -1 and end != -1 else ''


# Definition (sheet: Series)
def extract_definition(indicator_code):
    return series_dict_definition.get(indicator_code, '')


# Source (sheet: Series)
def extract_source(indicator_code):
    return data_dict_source.get(indicator_code, '')


# Subcategory
data['Subcategory'] = 'None'

# Link
data['Link'] = 'https://datatopics.worldbank.org/world-development-indicators/'

# Dataset
data['Dataset'] = 'WDI'

# Apply the functions to create columns
data['Category'] = data['Indicator Code'].apply(extract_category)
data['Indicator'] = data['Indicator Code'].apply(extract_indicator)
data['Dimension'] = data['Indicator Code'].apply(extract_dimension)
data['Units'] = data['Indicator Code'].apply(extract_units)
data['Definition'] = data['Indicator Code'].apply(extract_definition)
data['Source'] = data['Indicator Code'].apply(extract_source)

data = pd.concat([data[['Indicator Code', 'Category', 'Indicator', 'Dimension', 'Units', 'Definition', 'Source', 'Subcategory', 'Link', 'Dataset']]])

data.to_csv ('metadata_WDI.csv', index=False)
```
